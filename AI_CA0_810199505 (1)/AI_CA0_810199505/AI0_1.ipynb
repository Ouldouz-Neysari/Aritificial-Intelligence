جداسازی نجات یافته ها از غرق شده ها :

```python
rescued=_df[_df.Survived>1]
died=_df[_df.Survived<1]
```

میانگین و انحراف معیار تک تک ستون های نجات یافته ها را به دست می آوریم.

```python
rescued_means=[]
rescued_stds=[]
column_names=[]

for column in rescued:
  column_type=rescued[column].dtype
  if(column_type!=object):  
    this_mean=rescued[column].mean()
    this_std=rescued[column].std()
    rescued_means.append(this_mean)
    rescued_stds.append(this_std)
    column_names.append(column)



print(rescued_means)
print(rescued_stds)
```

میانگین و انحراف معیار تک تک ستون های غرق شده ها رو حساب می کنیم

```python
died_means=[]
died_stds=[]

for column in died:
  column_type=died[column].dtype
  if(column_type!=object):  
    this_mean=died[column].mean()
    this_std=died[column].std()
    died_means.append(this_mean)
    died_stds.append(this_std)



print(died_means)
print(died_stds)
```

```python
from scipy import stats
```

```python
fig, axs = plt.subplots(4,2)
xxx=np.linspace(-5,5,100)
for i in range(4):
  for j in range(2):
    if(i!=3 or j!=1):
      y_0_0=stats.norm.pdf(xxx,rescued_means[4*j+i],rescued_stds[4*j+i])
      y_0_1=stats.norm.pdf(xxx,died_means[4*j+i],died_stds[4*j+i])
      axs[i,j].plot(xxx,y_0_0)
      axs[i,j].plot(xxx,y_0_1)
      axs[i,j].set_title(column_names[4*j+i])
```

چون می خواهیم بر اساس داده جنسیت پیش بینی انجام دهیم ، ستون جنسیت را نرمال می کنیم با استفاده از میانگین و انحراف معیار داده های train :

```python

test_df['Sex'] = (test_df['Sex'] - _df['Sex'].mean()) / _df['Sex'].std()

```

```python
fig_spec,axs_spec=plt.subplots(1,1)

y_0_0=stats.norm.pdf(xxx,rescued_means[2],rescued_stds[2])
y_0_1=stats.norm.pdf(xxx,died_means[2],died_stds[2])
axs_spec.plot(xxx,y_0_0)
axs_spec.plot(xxx,y_0_1)
axs_spec.set_title(column_names[2])
```

```python
import math
std1=rescued_stds[2]
std2=died_stds[2]

mean1=rescued_means[2]
mean2=died_means[2]

A=((-1)/(std1**2))+((1)/(std2**2))

B=2*((mean1/(std1**2))-(mean2/(std2**2)))
C=((mean2**2)/(std2**2))-((mean1**2)/(std1**2))+math.log((std2**2)/(std1**2))

coefs=[A,B,C]
roooot=np.roots(coefs)
```

به این طریق توانستیم دو نقطه تقاطع را به دست آوریم.همان طور که از روی نمودار هم مشخص است بین این دو نقطه احتمال غرق شدن زیاد است برخلاف بیرون این دو نقطه که احتمال نجات بیشتر است.پس آرایه ای می سازیم و به ازای تک ردیف ها پیش بینی می کنیم که باتوجه به محدوده قرار گیری جنسیت نرمال سازی شده فرد نجات می یابد یا خیر .

```python
rescue_predicts=[]
for index, row in test_df.iterrows():
    if(row['Sex']<roooot[0] or row['Sex']>roooot[1]):
      rescue_predicts.append(1)
    else:
      rescue_predicts.append(0)
```

```python
predicted_df = test_df.assign(Survived=rescue_predicts)
predicted_df
```
