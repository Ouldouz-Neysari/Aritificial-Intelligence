q3)
Normalization or Standardization:

1)Normalization :
Normalization is used for rescaling our data to range of 0 to 1.It is used when the properties of data have diffrent and varying scales and we know that the scales of properties are not necassary for our prediction algorithms

2)Standardization :
Standardization is used for formating our data to standard normal distributation with 0 mean and 1 std.It is used when the properties of data have diffrent and varying scales and  our prediction algorithms assume that our data has GUASS distribution.

We are not sure about distribution of our data and we want to predict with KNN and... .So I think normalization can be better .

```python
normal_df=pandas.read_csv("diabetes.csv")

normal_df.fillna(normal_df.mode().iloc[0], inplace=True)
for col in list(normal_df.columns):

  df[col] = (df[col]-df[col].min())/(df[col].max()-df[col].min())

df
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

First We divide X , y ( the outcome of data frame) , then with train_test_split we split data in to train and test.

```python
X=normal_df.loc[:, normal_df.columns != "Outcome"]
y=normal_df["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)

clf1 = tree.DecisionTreeClassifier()
clf1 = clf1.fit(X_train, y_train)
predicted1 = clf1.predict(X_test)

predicted1

```

As we expect we take array of 0 and 1 for prediction 

We want to evaluate the model , so we use confusion martrix

```python
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(y_test,predicted1)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0,1])

cm_display.plot()
plt.show()
```

در این بخش حداکثر ارتفاع را کم کردیم تا تاثیر آن را روی دقت پیش بینی ببینیم 

در یک لوپ هر بار مقدارهای مختلفی به هایپر پارامتر حداکثر ارتفاع می دهیم و سپس با رسم ماتریکس آن تاثیر این تغییر را می بینیم .همچنین هر بار دقت مدل را هم به یک آرایه اضافه می کنیم :

```python

predictions_arr=[]
acc_arr=[]
for i in range(1,100,2):
  predicted_tmp=[]
  clf_tmp = tree.DecisionTreeClassifier(max_depth=i)
  clf_tmp = clf_tmp.fit(X_train, y_train)
  predicted_tmp = clf_tmp.predict(X_test)

  predictions_arr.append(predicted_tmp)
  acc_arr.append(metrics.accuracy_score(y_test, predicted_tmp))
  confusion_matrix_tmp = metrics.confusion_matrix(y_test,predicted_tmp)

  cm_display_tmp = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix_tmp, display_labels = [0,1])

  cm_display_tmp.plot()
  plt.show()


```

another hyperparameter of decision tree classifier is min_samples_split.This hyperparameter determine the minimum number of samples required to split an internal node. 

We create array of diffrent numbers and iterate over this array , each time we give one of this number as min_samples_split hyperparameter to the model

```python
min_sample_split = [2,3, 4,5, 6, 8, 10, 12, 14, 15, 16, 18, 20, 30]
```

```python
acc_arr2=[]
for ss in min_sample_split:
  predicted_tmp=[]
  clf_tmp = tree.DecisionTreeClassifier(min_samples_split=ss)
  clf_tmp = clf_tmp.fit(X_train, y_train)
  predicted_tmp = clf_tmp.predict(X_test)
  acc_arr2.append(metrics.accuracy_score(y_test, predicted_tmp))
  confusion_matrix_tmp = metrics.confusion_matrix(y_test,predicted_tmp)

  cm_display_tmp = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix_tmp, display_labels = [0,1])

  cm_display_tmp.plot()
  plt.show()
```

طبق ماتریکس ها می توانیم تاثیر کم شدن دقت با افزایش هایپر پارامتر را ببینیم . با افزایش این هایپرپارامتر تعداد نود هایی که مدل نیاز دارد تا آن ها را بسط دهد افزایش می یابد در نتیجه تعدادی از نود ها را که تعداد اعضا آن ها کمتر از مقدار تعیین شده است بسط نمیدهد.در اینجا هم می توانیم ببینیم که این هایپر پارامتر به تنهایی تعیین کننده نیست

##overfitting and underfitting

overfitting: Overfitting is an undesirable machine learning behavior that occurs when the machine learning model gives accurate predictions for training data but not for new data (high variance)

underfitting:Underfitting is a scenario in data science where a data model is unable to capture the relationship between the input and output variables accurately, generating a high error rate on both the training set and unseen data.(high bias)

ما در مدل هایمان سعی می کردیم با تنظیم کردن هایپرپارامتر ها از این دو مورد جلوگیری اما در حالاتی به آن ها برمیخوردیم . زمانی که از تابع برای تنظیم هایپر پارامتر ها استفاده می کردیم چون خوبی پارامتر ها بر اساس داده های آموزش سنجیده می شد گاهی به این حالت برمیخوردیم .

When a data engineer modifies the ML algorithm to better fit a given data set, it will lead to low bias—but it will increase variance. This way, the model will fit with the data set while increasing the chances of inaccurate predictions.

 Random forest takes more than one decision tree and this make it more accurate in comparsion with decision tree.
